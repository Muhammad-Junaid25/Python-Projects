```python
import pandas as pd
from scipy.stats import stats

# Selecting and performing theone way ANOVA test for Days of Week

# H0 : Number of accidents over each day is the same
# H1 : Number of accidents over each day are not the same

fvalue, pvalue = stats.f_oneway(df['speed_limit'] == 20, 
                                df['speed_limit'] == 30, 
                                df['speed_limit'] == 40, 
                                df['speed_limit'] == 50, 
                                df['speed_limit'] == 60, 
                                df['speed_limit'] == 70)


print("F Value  = {:g} ".format(fvalue))
print("P Value  = {:g} ".format(pvalue))

alpha = 0.05

if pvalue < alpha:  # null hypothesis: x comes from a normal distribution
    print("The null hypothesis is rejected")
else:
    print("The null hypothesis is accepted")


```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import preprocessing

import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import utils
from tensorflow.keras.layers import Dense

# Import label encoder
from sklearn import preprocessing


# Select the features and target variable
X = df[['speed_limit']]
y = df['accident_severity']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=99)

# Preprocess the data
scaler = StandardScaler()

# Fit only to the training data
scaler.fit(X_train)

# Now apply the transformations to the data:
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

# Label encode the target variable
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in columns.
y_train_nn = label_encoder.fit_transform(y_train)
y_test_nn = label_encoder.fit_transform(y_test)

# Normalize the input data
x_train_ann = X_train.astype("float32") / 255
x_test_ann = X_test.astype("float32") / 255

num_classes = 3
y_train_ann = utils.to_categorical(y_train_nn, num_classes)
y_test_ann = utils.to_categorical(y_test_nn, num_classes)
num_features = x_test_ann.shape[1]
num_classes = y_train_ann.shape[1]

# Define the model using TensorFlow Keras
model = tf.keras.Sequential()
model.add(Dense(512, kernel_initializer='normal', activation='relu', input_shape=(num_features,)))
model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
history = model.fit(x_train_ann, y_train_ann, validation_data=(x_test_ann, y_test_ann), 
                    epochs=20, batch_size=200, verbose=1)


# Evaluate the model on the test data
score = model.evaluate(x_test_ann, y_test_ann, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# Get the predicted classes for the test data
y_pred = model.predict(x_test_ann)

# Print the classification report
y_pred_nn = np.argmax(y_pred, axis=1)
RP = classification_report(y_test_nn, y_pred_nn)

print(RP)

# create confusion matrix# create confusion matrix
matrix = confusion_matrix(y_test_nn, y_pred_nn)

# create heatmap
sns.heatmap(confusion_matrix(y_test_nn, y_pred_nn), annot=True, cbar=None, cmap='RdGy')
plt.title('Confusion Matrix')
plt.tight_layout(), plt.xlabel('True Values'), plt.ylabel('Predicted Values')
plt.show()
```
